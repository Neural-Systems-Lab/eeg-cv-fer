```python
from fer import FER
import matplotlib.pyplot as plt

DATA_PATH = "img-data/"
```

```python
def detect_still_img(img_path):
    img = plt.imread(DATA_PATH + img_path)
    detector = FER(mtcnn=True)
    print(detector.detect_emotions(img))
    plt.imshow(img)
```

```python
detect_still_img("confused-josh.jpg")
```
